# Plotting from a Jupyter notebook

```python
# %matplotlib command - are two optiooptions
# %matplotlib notebook – This command will produce interactive plots embedded within the notebook.
# %matplotlib inline – It will output static images of the plot embedded in the notebook.
```

```python
%matplotlib inline

x1 = np.linspace(0, 10, 100)
fig = plt.figure()               # create a plot figure

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--');
plt.show()
```

```python
# create a plot figure 
plt.figure()

#create the first of two panels and set current axis
plt.subplot(2, 1, 1)         # (rows, columns, panel number)
plt.plot(x1,np.sin(x1));

#create the second of two panels and set current axis
plt.subplot(2, 1, 2)         #(rows, columns, panel number)
plt.plot(x1, np.cos(x1));
plt.show()
```

# State-machine interface


```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

# Working with Numpy arrays

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

# Object-Oriented API

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-');                      # by removing the " ; " at commond sin 
plt.show()
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'b-');                # adding the " ; " to the commond
ax[1].plot(x1, np.cos(x1), 'b-');
plt.show()
```

# Objects and Reference 

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2');
axes.set_title('title');
plt.show()
```

# Error Bar Chart

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show();
```

# Stacked Bar Chart

```python
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

# Adding a legend

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0);

ax.legend(['Normal','Fast','Slow']);
plt.show()
```

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow');

ax.legend();
plt.show()
```
