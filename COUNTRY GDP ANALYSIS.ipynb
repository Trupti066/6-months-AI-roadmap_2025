# We analyse python dataframe or dataset

```python
# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

import matplotlib.pyplot as plt # visulaiztion
import seaborn as sns # distribution visualtion 

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

#import warnings
#warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
df["InternetUsers"]
```

```python
#distributions:
visl = sns.histplot(df["InternetUsers"])
#plt.show(Visl)
plt.show()
```

```python
#distributions:
visl = sns.distplot(df["InternetUsers"])         
# plt.show(Visl)
plt.show()
```

```python
vis2 = sns.displot(df["InternetUsers"]);
plt.show()
```

```python
vis2 = sns.displot(df["InternetUsers"],bins=15);
plt.show()
```

```python
# Box Plot:
vis4 = sns.boxplot(data = df, x="IncomeGroup", y='BirthRate')     #Bi-variate analysis
plt.show()
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate');
plt.show()
```

```python
vin8 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup')             #hue - parameter for color
plt.show()
```

```python
vvis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                   fit_reg = True,hue = 'IncomeGroup');
plt.show()
```
